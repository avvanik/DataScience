```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def encode_df(dataframe):
    le = LabelEncoder()
    for column in dataframe.columns:
        dataframe[column] = le.fit_transform(dataframe[column])
    return dataframe

encode_df(df)
```

```python
X = df.iloc[:, :4]
y = df.iloc[:,-1:]

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
from sklearn.tree import export_graphviz

features = df.columns
features = features[:-1]
class_names = list(set(str(df.iloc[:,-1])))

export_graphviz(
         tree_clf,
     out_file="D:\GDRIVE\DS-DATA\JUPYTER-NOTEBOOKS\IRIS-DecTree\iris_tree.dot",
         feature_names=features,
         class_names=class_names,
         rounded=True,
         filled=True
)
```
