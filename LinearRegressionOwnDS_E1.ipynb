```python
import numpy
import matplotlib.pyplot as plot
from scipy import stats
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/avvanik/DataScience/main/w.history.csv')
df.head()
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
# print(x)
# print(y)
stats.describe(y)
plot.plot(y)
plot.show()
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 1/7, random_state = 0)
print(xTrain)
print(yTrain)
```

```python
linearRegressor = LinearRegression()
linearRegressor.fit(xTrain, yTrain)
print(linearRegressor)
yPrediction = linearRegressor.predict(xTest)
print(yPrediction)
```

```python
plot.scatter(xTrain, yTrain, color = 'red')
plot.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plot.title('days vs screentime (Training set)')
plot.xlabel('days')
plot.ylabel('screentime in minutes')
plot.show()
plot.scatter(xTest, yTest, color = 'red')
plot.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plot.title('days vs screentime (Test set)')
plot.xlabel('days')
plot.ylabel('screentime in minutes')
plot.show()
```
